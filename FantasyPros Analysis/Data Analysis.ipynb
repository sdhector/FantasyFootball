```python
from DataExtraction import PlayerArchive 
from DataExtraction import RosterData 
from DataExtraction import PlayerData 
```

```python
PlayerArchive = PlayerArchive()
RosterTable = RosterData()
PlayerData_2019, TeamData_2019 = PlayerData()
```

```python
temp = (PlayerData_2019[PlayerData_2019.Pos.isnull()]
        .merge(PlayerArchive, 
               left_on = 'Player', 
               right_on = 'Name', 
               how = 'inner')
        [['Player', 'TeamName','Position', 'CareerStart', 'CareerEnd']]
        .drop_duplicates()
        .sort_values(['Player', 'TeamName', 'CareerEnd'], ascending = [1,1,0])
        .groupby(['Player', 'TeamName'])
        .agg('first')
        .reset_index())[['Player','TeamName', 'Position']]
temp = PlayerData_2019.merge(temp, on = ['Player', 'TeamName'], how = 'left')
PlayerData_2019.loc[temp.Pos.isnull(),'Pos'] = temp[temp.Pos.isnull()].Position
PlayerData_2019.head(5)
```
