# NFL Stating out

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
```

## Download the URL Page Content

```python
url = 'https://www.pro-football-reference.com/teams/mia/2019.htm'
driver = webdriver.Chrome()
driver.get(url)
page_source = driver.page_source
#r = requests.get(url)
#html = r.text.
soup = BeautifulSoup(page_source, 'html.parser')
driver.quit()
print(soup.prettify())
```

## Getting Test Table: Total Team Stats

```python
TableRows = (soup.find('div', {'id':'all_team_stats'})
        .find('table',{'class':'sortable stats_table'})
        .find('tbody')
        .find_all('tr')
       )
TableColumn_Name = []
TableColumn_Value = []
FirstColumn = []
for j in range(0,len(TableRows)):
    FirstColumn.append(TableRows[j].find_all('th')[0].get_text())
    TableColumns = TableRows[j].find_all('td')
    listA = []
    listB = []
    for i in range(0,len(TableColumns)):
        listA.append(TableColumns[i].get_text())
        listB.append(TableColumns[i].get_attribute_list('data-stat')[0])
    
    TableColumn_Value.append(listA)
    TableColumn_Name = listB

TestTable = pd.concat(
    [pd.DataFrame(FirstColumn, columns = ['player']), 
     pd.DataFrame(TableColumn_Value, columns = TableColumn_Name)], 
    axis = 1 )
TestTable.head()
```
